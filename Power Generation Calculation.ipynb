# Surface Area of Wind Turbine


### A - Surface Area of the Turbine Propeller
### P - Total Power output desired - ex. 50MW
### $\rho$ - air density 1.255 $\frac{kg}{m^3}$
### v - wind speed based on hurricane strength
##### Category 1: 119 - 153 Km/h
##### Category 2: 154 - 177 Km/h
##### Category 3: 178 - 208 Km/h
##### Category 4: 209 - 251 Km/h
##### Category 5: 252 & up Km/h
### e - efficiency of turbine


    

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
air_density = 1.255 #kg/m^3
goal_power = 50.*10**6 #W/h
wind_speed = np.linspace(100*1000/3600.,280*1000/3600.) #m/s
efficiency = 0.45
num_turbines = 1 #if you want to check number of turbines needed
A = 2*goal_power/(wind_speed**3*efficiency*air_density*num_turbines)
plt.plot(wind_speed*3600/1000,A)
plt.xlabel(r'Wind Speed ($km/h$)')
plt.ylabel(r'Blade Surface Area ($m^2$) ')
# plt.title('Total With {} Turbines and  {}% Efficiency'.format(num_turbines,efficiency))
# plt.title('Surface Area of Turbines and {}% Efficiency'.format(efficiency))
plt.title('Surface Area and {}% Efficiency of Turbines to Produce 50MW/h'.format(efficiency))


plt.axvline(178,color='red')
plt.axvline(208,color='black')
# plt.axvline(209,color='green')
plt.axvline(251,color='green')
# plt.xlim(175,260)
# plt.ylim(0,2000)
plt.savefig('turbineSA.png',dpi=300)
```
